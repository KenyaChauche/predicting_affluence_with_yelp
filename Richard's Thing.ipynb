```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_slice(df, y, xlabels):
    fig, ax = plt.subplots(nrows=3, ncols=1, sharex = True, figsize = (4, 12))
    ax = ax.ravel()
    for i, l in enumerate(xlabels):
        ax[i].scatter(df[l], y)
        ax[i].set_title(l)
    plt.show()
```

```python
df = pd.read_csv("../datasets/generated_gps_price_radius.csv")
```

```python
df.head()
```

```python
plot_slice(df, y = df["median rent"], xlabels = ["0.5mi 1 dollar", "0.5mi 2 dollar", "0.5mi 3 dollar"])
```
