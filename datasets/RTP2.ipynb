```python
import requests
from geopy.distance import geodesic
key = 
def yelp(N, center, radius, key):
    D = {}
    url = 'https://api.yelp.com/v3/businesses/search'
    headers = {'Authorization': 'bearer %s' % key}
    for i in range(N // 50):
        params = {'limit' : 50, 'offset' : 50 * i, 'radius' : radius, 'latitude' : center[1], 'longitude' : center[0]}
        R = requests.get(url = url, params = params, headers = headers).json()
        if not 'businesses' in R:
            break
        R = R['businesses']
        for B in R:
            try:
                W = {'alias' : B['alias'], 'latitude' : B['coordinates']['latitude'], 'longitude' : B['coordinates']['longitude'], 'price' : len(B['price'])}
                if geodesic((W['latitude'], W['longitude']), center[::-1]).meters <= radius:
                    D[B['id']] = W
            except:
                pass
    return D
def sample(N, L, radius):
    D = {}
    for z in L:
        Y = yelp(N, z, radius, key)
        D[z] = {i : sum(Y[k]['price'] == i for k in Y) for i in range(1, 5)}
    return D
```
